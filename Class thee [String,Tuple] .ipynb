# Tuple

**Tuple**
1. slicing supports
2. Indexing suppports
3. Immutable
4. Cannot be modified
  - no add, remove or item assigning

```python
tuple1 = (1, 'ad', 'one', 2.4)
tuple1[0]
```

**Tupple suppports list**

```python
tuple2 = (1,[1,2,3,4],'hello')
print(tuple2)
print(tuple2[1][1])
tuple2[1][1] = 50
print(tuple2)
```

**We can use tupple in tupple**

```python
tuple3 = (1,2,(3,4,5),'israq', 8.9)
print(tuple3)
tuple3[2][1]=7
```

**Converting tuple into list**

```python
print('\n At first tuple3 was like:', tuple3)
tuple3 = list(tuple3)
print('\n After converting tuple3 into list:', tuple3)
tuple3[1] = 60
print('\n Now we can change the values of tuple' ,tuple3)

tuple3[2] = list(tuple3[2])
print('\n Now we can change the inner tuple into lists',tuple3)

tuple3[2] = tuple(tuple3[2])
print('\n Now we again change list into tuple:',tuple3)

tuple3 = tuple(tuple3)
print('\n Again changing the whole list into tuple',tuple3)

```
