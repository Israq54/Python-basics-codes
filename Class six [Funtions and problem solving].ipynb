# Code for a prime number

```python
x = int(input("Enter your number: "))
print("You have entered", x)
y = 0
for i in range(2,x-1):
  if x%i == 0:
    print("It is not a prime number.")
    y = 1
    break
if y == 0:
  print("It is a prime number.")
```
