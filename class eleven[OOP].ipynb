# Constructors

```python
# constructor
class sample:
  def __init__(self,name):
    self.name = name
  def disp(self):
    print(self.name)

obj = sample('israq')
obj.disp()
```

```python
class sample:
  def __init__(self):
    name = input()
    self.name = name
  def disp(self):
    print(self.name)

obj = sample()
obj.disp()
```

```python
from abc import ABC,abstractmethod
class sample(ABC):
  @abstractmethod
  def abstract_method(self):
    pass

class child(sample):
  def abstract_method(self):
    print("abstraction")

obj = child()
obj.abstract_method()
```

```python
from abc import ABC,abstractmethod
class sample(ABC):
  @abstractmethod
  def print(self):
    print('abstract method')

class child(sample):
  def print(self):
    print('child class')

obj = child()
obj.print()
```

```python
from abc import ABC,abstractmethod
class sample(ABC):
  @abstractmethod
  def print(self):
    print('abstract method')

class child(sample):
  def print(self):
    super().print()
    print('child class')

obj = child()
obj.print()
```
