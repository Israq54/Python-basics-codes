# 7. Write a python program to find the first appearance of the substring'not' and 'poor' in a given string. If 'not' follows 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string.

```python
s1 = 'The lyric is not that poor.'

snot = s1.find('not')
spoor = s1.find('poor')

if spoor > snot and snot>0 and spoor> 0:
  s1 = s1.replace(s1[snot:spoor+4], 'good')
  print(s1)
else:
  print(s1)
```

# 12. Write a python program to count the occurances of each word in a given sentence.

```python
s1 = 'the quick fox jumps over the lazy dog.'

s1 = s1.split()
counts = dict()

for i in s1:
  if i in counts:
    counts[i] = counts[i] + 1
  else:
    counts[i] = 1
print(counts)
```
