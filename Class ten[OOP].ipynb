```python
#multilevel inheritance
class parent:
  def __init__(self):
    self.name = "israq"

class child(parent):
  def __init__(self):
    parent.__init__(self)
    self.age = 40

class child2(child):
  def __init__(self):
    child.__init__(self)
    self.prof = "banker"
  def display(self):
    print(self.name,self.age,self.prof)
obj = child2()
obj.display()
```

```python
#multiple inhertence
class parent1:
  def __init__(self):
    self.name = "john"

class parent2:
  def __init__(self):
    self.age = 40

class child(parent1,parent2):
  def __init__(self):
    self.prof = "banker"
    parent2.__init__(self)
    parent1.__init__(self)

  def display(self):
   print(self.name, self.age,self.prof)

obj = child()
obj.display()
```

```python
#hierarchical inheritance
class parent:
  def __init__(self):
    self.name = "israq"
    self.age = 40
    self.prof = "banker"

class child1(parent):
  def __init__(self):
    self.city = "paris"
    parent.__init__(self)
  def display(self):
    print(self.name,self.age,self.city)

obj = child1()
obj.display()

obj2 = child2()
obj2.display()
```
