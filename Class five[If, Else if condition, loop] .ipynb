# If, else if condition

**Checking who is bigger using if and elif condition**

```python
a = 25
b = 23

if a > b:
  print('A is big.')
elif b > a:
  print('B is big.')
elif b==a:
  print("Both are same")
```

Check whether element in list is even or odd. If it is even -  square or else divide by 2

```python
print(list1)

for i in range(0,len(list1)):
  if list1[i]%2 == 0:
    print(list1[i]**2)
  else:
    print(list1[i]/2)
```

# While loop

Comparing between for and while
