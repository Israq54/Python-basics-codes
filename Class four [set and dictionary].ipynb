# Set

1. We will use curly brackets {} for set
2. In set indexing and slinging doesn't support
3. Set doesn't take same value more than once
4. Set maintains serial
5. Set is mutable
6. while using union the value won't be updated we will be needed to assign
7. while using update the value will automatically updated

```python
set1 = {1,2}
type(set1)
```

**Set doesn't take one value more than once**

**Set maintains serial**

```python
set2 = {1,3,2,4,4,5,5,6,6,6}
list2 = [1,3,2,4,4,5,5,6,6,6]

print(set2)
print(list2)
```

**Add and remove**

```python
set3 = {1,2,3,4}
set4 = {7,8,9}

set3.add(100)
print(set4)

set3.remove(4)
print(set4)
```

**Intersection in set**

```python
set3 = {1,2,3,4}
set4 = {7,8,9,1,2}

set3 = set3.intersection(set4)
print(set3)
```

**Union in set**

```python
set3 = {1,2,3,4}
set4 = {7,8,9,1,2}

set3 = set3.union(set4)
print(set3)
```

**SetA - SetB**

```python
set3 = {1,2,3,4}
set4 = {7,8,9,1,2}

set3 = set3.difference(set4)
print(set3)

set3 = {1,2,3,4}

set4 = set4.difference(set3)
print(set4)
```

**Dictionay in a Dictionary**

```python
classroom = {'girls': {'g1': 90, 'g2': 78, 'g3': 45},
            'boys':{ 'b1': 67, 'b2': 89, 'b3': 67}}
```

Get Specific values in a complex dictionary

```python
print(classroom['girls']['g3'])
print(classroom['boys']['b2'])
```

**Adding new word in dictionary**

```python
print('Before adding words in dictionary: \n', classroom)

classroom['foriegner'] = [{'Indian': '5', 'nepali': '35', 'Bhutan': '44'}]

print("\nAfter adding words in dictionary:\n", classroom)
```

**Deleting values in dictionary**

```python
del classroom['girls']
print(classroom)
```

**Add to dictionary**

```python
classroom.update(dict2)
print(classroom)
```

**Changing values of a keys in dictionary**

```python
classroom['boys']['b3'] = '68'

print(classroom)
```

**Changing keys in a dictionary**

```python
dict2['w11'] = dict2['w1']
print(dict2)
del dict2['w1']
print(dict2)
```
